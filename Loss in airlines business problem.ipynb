```python

# Merging Booking Data with Ticket Data to know how many tickets are booked  with respect to each booking reference

bookings_ticket_df=pd.merge(bookings,tickets, on='book_ref', how='inner')

# printing bookings_ticket_df 

bookings_ticket_df

```

```python
# Finding Duplicate Rows 

duplicate =bookings_ticket_df[bookings_ticket_df .duplicated()]
 
print("Duplicate Rows :")
 
```

```python
# Merging bookings_ticket_df with ticket_flights to know what is the ticket_no,flight_no,fare_condition and amount paid for that flight

bookings_ticket_ticket_flights_df=pd.merge(bookings_ticket_df,ticket_flights, on='ticket_no', how='inner')


# printing bookings_ticket_ticket_flights_df


bookings_ticket_ticket_flights_df.head(5)
```

```python
#  scheduled_departure dekhane k liye ki year month day and hour sahi split hue  

#df3[['scheduled_departure','year','month','day','hour']]
```

```python
# Creating a new column Travel_Date which has year month and day

df3["Travel_Date"]=pd.to_datetime(df3[["year", "month", "day"]])
```

```python
df3.head(5)
```

```python
# Grouping df3 data frame with  'Travel_Date', 'hour', 'flight_no', 'aircraft_code', 'route', and 'fare_conditions'  to get the information that which flight number travelled on which date at what time on which route by which aircraft and with what fare condition

Grouped_df=df3.groupby(['Travel_Date','hour','flight_no','aircraft_code','route','fare_conditions'])
```

```python
# Printing Grouped_df data frame

Grouped_df.head()
```

```python
# Finding the sum of the grouped data frame

Grouped_df_Sum=Grouped_df.sum()


Grouped_df_Sum

```

```python
# Finding the count of grouped data frame

Grouped_df_Count=Grouped_df.count()
```

```python
# Printing Grouped_df_Count data frame

Grouped_df_Count
```

```python
# Exploring seats

seats
```

```python
# Making a copy of seats

seats_info=seats.copy()
```

```python
# Grouping seats by 'aircraft_code','fare_conditions'

seats_info_merge= seats_info.groupby(['aircraft_code','fare_conditions'])
 
```

```python
# Calculating the count of Business and Economy seats in each aircraft

seat_agg=seats_info_merge.count()
```

```python
# Printing seat_agg

seat_agg
```

```python
#Joining  Grouped_df_Count with seat_agg to find the total number of seats each aircraft is having

Grouped_df_Count_withoccpancy = Grouped_df_Count.join(seat_agg,on=['aircraft_code','fare_conditions'], how='left', lsuffix='_left', rsuffix='_right')

```

```python
# Printing Grouped_df_Count_withoccpancy

Grouped_df_Count_withoccpancy
```

```python
# Finding occupancy of each flight by dividing the number of seats filled to the total number of available seats


Grouped_df_Count_withoccpancy["occupancy"]=Grouped_df_Count_withoccpancy["seat_no_left"]/Grouped_df_Count_withoccpancy["seat_no_right"]
```

```python
# Printing Grouped_df_Count_withoccpancy data frame

Grouped_df_Count_withoccpancy
```

```python
# Printing Grouped_df_Sum

Grouped_df_Sum
```

```python
# Printing Grouped_df_Count_withoccpancy

Grouped_df_Count_withoccpancy
```

```python
#Combining Grouped_df_Sum with Grouped_df_Count_withoccpancy to know what amount each flight earn and what is its occupancy


Grouped_df_Count_with_amount_and_occupancy=Grouped_df_Sum.join(Grouped_df_Count_withoccpancy,on=['Travel_Date','hour','flight_no','aircraft_code','route','fare_conditions'], how='left', lsuffix='_left', rsuffix='_right')

Grouped_df_Count_with_amount_and_occupancy
```

```python
# We will export this data frame in Grouped_df_Count_with_amount_and_occupancyexcel sheet and will do further analysis on excel sheet

# determining the name of the file

file_name1 = 'Grouped_df_Count_with_amount_and_occupancy.xlsx'
  
# saving the excel

Grouped_df_Count_with_amount_and_occupancy.to_excel(file_name1)
#print('DataFrame is written to Excel File successfully.')
```
